```python
def snap_map(results):
    snap_lst = []
    for i in range(0, len(results)):
        loc = str(results[i]['location']['latitude'])+','+str(results[i]['location']['longitude'])
        snap_lst.append(loc)
    aaa = pd.DataFrame({'location':snap_lst})
    return jay.google_map(aaa)
```
